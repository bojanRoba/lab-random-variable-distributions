```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_column',None)
```

```python
# 1.Use the numerical variables from our dataset this week.

customer_num = pd.read_csv(r'..\lab-cleaning-numerical-data\numerical.csv')
customer_num
```

```python
# 2.Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

for colm in customer_num:
    sns.distplot(customer_num[colm])
    plt.show()
```

**customer lifetime value** - the plot shows that the CLV distribution is skewed to the right, with most customers having a CLV of less than 20,000 USD and a few customers having a CLV of between 20,000 and 80,000 USD.

**monthly_premium_auto** - the plot shows that the distribution is skewed to the right, with most subscriptions costing less than 150 USD per month and a few subscriptions costing more than 150 USD per month.

**total_calim_amount** - the plot shows that the distribution is skewed to the right, with most claims being for relatively small amounts (less thana 1,000 USD) and a few claims being for higher then 1,000 USD.

```python
# 3.Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

customer_num.apply(np.log)
```

```python
# the problem with log transformation is that since some of the observations are zero, 
# then the log will give a -inf value. to overcome this, I will add a very small number to all observations 
# before the log transformation and therefore do not have zero values.

def log_update(x):
    
    output = np.log(x + 0.001)
    return output

customer_num_log = customer_num.apply(log_update)
customer_num_log
```

```python
# plotting the dist

for x in customer_num_log.columns:
    sns.distplot(customer_num_log[x], kde=True)
    plt.show()
```
